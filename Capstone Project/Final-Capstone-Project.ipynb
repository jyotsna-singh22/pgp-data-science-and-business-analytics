```python
# Outlier treatment
def detect_outliers(col):
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
sns.boxplot(insurance_df['daily_avg_steps'])
```

```python
lr,ur=detect_outliers(insurance_df['daily_avg_steps'])
print('lower range- ',lr,'upper range- ',ur)
```

```python
insurance_df['daily_avg_steps']=np.where(insurance_df['daily_avg_steps']>ur,ur,insurance_df['daily_avg_steps'])
insurance_df['daily_avg_steps']=np.where(insurance_df['daily_avg_steps']<lr,lr,insurance_df['daily_avg_steps'])
```

```python
sns.boxplot(insurance_df['daily_avg_steps'])
```

```python
import math
current_year = float(2022)
def year_diff(year):
    if math.isnan(year):
        return 'nan'
    else:
        return current_year - year
```

```python
# Adding a new column 
year_since_last_admitted=[]
for x in df['Year_last_admitted']:
    year_since_last_admitted.append(year_diff(x))
    
df.insert(loc=21,column='years_since_last_admitted',value=year_since_last_admitted)

```

# Question 4 - Model building

```python
# Sparating dependent and independent variables into x and y
x=df.drop('insurance_cost',axis=1)
y=df.pop('insurance_cost')
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting data into train and test set in 30:70 ratio
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=1)
```

# 1. Random forest model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Creatin random forest classifier model
rf= RandomForestRegressor(n_estimators=501,random_state=1)
```

```python
# Fitting train set into the model
rf.fit(x_train,y_train)
```

```python
# Validating model on train set
rf_train_predict=rf.predict(x_train)
```

```python
from sklearn import metrics
```

```python
print('rf_train_RMSE : ',np.sqrt(metrics.mean_squared_error(y_train,rf_train_predict)))
```

```python
# Defining a function to calculate MAPE
def MAPE(y_actual,y_predicted):
    mape=np.mean(np.abs((y_actual-y_predicted)/y_actual))*100
    return mape
```

```python
# Calculating MAPE
rf_train_MAPE=MAPE(y_train,rf_train_predict)
print('rf_train_MAPE : ',rf_train_MAPE)
```

```python
rf_train_adj_r2=1-(1-metrics.r2_score(y_train,rf_train_predict))*(len(df)-1)/(len(df)-(df.shape[1]-1)-1)
print('rf_train_adjr2 : ',rf_train_adj_r2)
```

```python
# Validating on test set 
rf_test_predict=rf.predict(x_test)
```

```python
print('rf_test_RMSE : ',np.sqrt(metrics.mean_squared_error(y_test,rf_test_predict)))
rf_test_MAPE=MAPE(y_test,rf_test_predict)
print('rf_test_MAPE : ',rf_test_MAPE)
rf_test_adj_r2=1-(1-metrics.r2_score(y_test,rf_test_predict))*(len(df)-1)/(len(df)-(df.shape[1]-1)-1)
print('rf_test_adjr2 : ',rf_test_adj_r2)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Tuning the model
param_grid={'max_depth':[10],
           'max_features':[15],
           'min_samples_leaf':[5],
          'min_samples_split':[10],
           'n_estimators':[501]}
```
